```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
sns.get_dataset_names()
```

```python
penguins=sns.load_dataset('penguins')
```

```python
penguins.head()
```

```python
penguins['species'].value_counts()
```

```python
penguins['island'].value_counts()
```

```python
sns.scatterplot(data=penguins,x='flipper_length_mm',y='body_mass_g',hue='island',palette='Set1',style='sex')
sns.set_style('ticks')
sns.despine()
```

```python
sns.stripplot(data=penguins,x='species',y='body_mass_g',hue='island',dodge=True,jitter=True)
sns.despine()
```

```python
sns.swarmplot(data=penguins,x='species',y='body_mass_g',hue='island',dodge=True)
sns.despine()

sns.set_context('notebook')
```

```python
sns.set_context('notebook')
sns.set_style('whitegrid')
sns.histplot(data=penguins,x='body_mass_g',hue='sex',multiple="stack",)
```

```python
sns.set_context('notebook')
sns.set_style('whitegrid')
sns.lineplot(data=penguins,x='body_mass_g',y='flipper_length_mm',hue='island',style='sex')
```

```python
sns.jointplot(data=penguins,x='body_mass_g',y='flipper_length_mm')
```

```python
sns.jointplot(data=penguins,x='body_mass_g',y='flipper_length_mm',hue='sex',kind="kde")
```

```python
sns.jointplot(data=penguins,x='body_mass_g',y='flipper_length_mm',hue='sex',kind="scatter")
```

```python
sns.barplot(data=penguins,x="species",y="body_mass_g",hue="sex",palette=['pink','skyblue'],estimator=np.sum)
```

```python
sns.countplot(data=penguins,x='species',hue='island',palette='Set2')
```

```python
sns.boxplot(data=penguins,x='species',y='body_mass_g',hue='sex')
```

```python
columns=["bill_length_mm","bill_depth_mm","flipper_length_mm","body_mass_g"]
penguins[columns].corr
sns.heatmap(data=penguins[columns].corr(),annot=True,cmap='coolwarm')
```

```python

sns.heatmap(data=penguins[columns].corr(),annot=True,cmap='Blues',vmin=-0.2,linewidths="3",linecolor='black')
```

```python
sns.rugplot(data=penguins,y="body_mass_g",x='species',palette='Set2',height=0.2)
```
