```python
df = pd.DataFrame({"Name": ["Alice", "Bob"], "Marks": [50, 80]})


def grade(marks):
    if marks >= 75:
        return "A"
    elif marks >= 50:
        return "B"
    else:
        return "C"
    
df["Grades"]=df["Marks"].apply(grade)
df
```

```python
def name_marks(row):
    return f"{row['Name']} scored {row['Marks']}"
df['Summary']=df.apply(name_marks,axis=1)
df
```
